# **THEORY QUESTION**

# Question 1: What is hypothesis testing in statistics?

 -  Hypothesis testing is a statistical method used to make decisions about a population parameter based on sample data.
It involves formulating a null hypothesis (H₀) and an alternative hypothesis (H₁), then using statistical tests to
determine whether to reject or fail to reject the null hypothesis based on the evidence from the sample.


# Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

 -  The null hypothesis (H₀) is a statement of no effect or no difference that serves as the default assumption.
The alternative hypothesis (H₁) is what the researcher wants to prove, representing a significant effect or difference.

Key differences:
- H₀ assumes no change/effect/difference, while H₁ suggests there is one
- H₀ is what we test against, H₁ is what we hope to support
- Rejecting H₀ provides evidence for H₁


# Question 3: What is the significance level in hypothesis testing, and why is it important?

 -  The significance level (α) is the probability threshold for rejecting the null hypothesis, typically set at 0.05 or 5%.
It represents the maximum acceptable probability of making a Type I error (falsely rejecting H₀ when it's true).

Importance:
- Determines the critical region for test statistics
- Controls the balance between Type I and Type II errors
- Provides a standard for decision making in statistical tests


# Question 4: What does a P-value represent in hypothesis testing?

  - The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis is true.

Interpretation:
- Small P-value (< α): Strong evidence against H₀ (reject H₀)
- Large P-value (≥ α): Weak evidence against H₀ (fail to reject H₀)


# Question 5: How do you interpret the P-value in hypothesis testing?

 -  P-value interpretation:
- P < 0.01: Very strong evidence against H₀
- 0.01 ≤ P < 0.05: Strong evidence against H₀
- 0.05 ≤ P < 0.10: Weak evidence against H₀
- P ≥ 0.10: Little or no evidence against H₀

Note: It's not the probability that H₀ is true, but rather how incompatible the data are with H₀.


# Question 6: What are Type I and Type 2 errors in hypothesis testing?

 -  Type I error (False Positive): Rejecting H₀ when it's actually true
- Probability = significance level (α)

Type II error (False Negative): Failing to reject H₀ when it's actually false
- Probability = β

The power of a test is 1-β, the probability of correctly rejecting H₀ when it's false.


# Question 7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

 -  One-tailed test:
- Tests for effect in one direction only (either greater than or less than)
- More powerful for detecting effects in the specified direction
- Used when we have a specific directional hypothesis

Two-tailed test:
- Tests for effect in both directions
- More conservative
- Used when we're interested in any difference from H₀


# Question 8: What is the Z-test, and when is it used in hypothesis testing?

 - The Z-test is a parametric statistical test that compares sample means to a population mean when:
1. The population standard deviation is known
2. The sample size is large (typically n > 30)
3. The data follows a normal distribution

Common uses:
- Testing if a sample mean differs from a known population mean
- Comparing two sample means
- Testing proportions


# Question 9: How do you calculate the Z-score, and what does it represent in hypothesis testing?

 -  Z-score formula:
Z = (X̄ - μ) / (σ/√n)
Where:
X̄ = sample mean
μ = population mean
σ = population standard deviation
n = sample size

It represents:
- How many standard deviations the sample mean is from the population mean
- The test statistic in a Z-test
- Used to calculate P-values and make decisions about H₀


# Question 10: What is the T-distribution, and when should it be used instead of the normal distribution?

 -  The T-distribution is a probability distribution similar to the normal distribution but with heavier tails. It's used when:
- The sample size is small (typically n < 30)
- The population standard deviation is unknown
- The data is approximately normally distributed

Key differences from normal distribution:
- Shape depends on degrees of freedom (df = n-1)
- More probability in the tails
- Approaches normal distribution as df increases


# Question 11: What is the difference between a Z-test and a T-test?

 -  Z-test:
- Uses population standard deviation (σ)
- Uses normal distribution
- For large samples or known σ
- Z = (X̄ - μ) / (σ/√n)

T-test:
- Uses sample standard deviation (s)
- Uses t-distribution
- For small samples or unknown σ
- t = (X̄ - μ) / (s/√n)


# Question 12: What is the T-test, and how is it used in hypothesis testing?

-  The T-test is a statistical test used to compare means when:
1. Population standard deviation is unknown
2. Sample size is small
3. Data is approximately normal

Types:
1. One-sample t-test: Compare sample mean to known value
2. Independent samples t-test: Compare means of two independent groups
3. Paired t-test: Compare means from the same group at different times

Process:
1. Calculate t-statistic
2. Determine degrees of freedom (n-1)
3. Find critical t-value or P-value
4. Make decision about H₀


# Question 13: What is the relationship between Z-test and T-test in hypothesis testing?

 - Relationship:
- Both are used to test hypotheses about means
- Both calculate standardized distances between observed and expected values
- T-test becomes equivalent to Z-test as sample size increases (df → ∞)
- Z-test is a special case of t-test when σ is known

When to use which:
- Use Z-test when σ is known or n is large
- Use T-test when σ is unknown and n is small


# Question 14: What is a confidence interval, and how is it used to interpret statistical results?

 - A confidence interval is a range of values that likely contains the true population parameter with a specified probability (confidence level, typically 95%).

Interpretation:
- "We are 95% confident that the true parameter lies between X and Y"
- Wider intervals indicate more uncertainty
- If CI includes null value (e.g., 0 for differences), result is not statistically significant

Calculation for mean:
CI = X̄ ± (critical value * SE)
Where SE = standard error


# Question 15: What is the margin of error, and how does it affect the confidence interval?

 - The margin of error (ME) is half the width of the confidence interval, representing the maximum expected difference between the sample statistic and population parameter.

ME depends on:
1. Confidence level (higher level → larger ME)
2. Sample size (larger n → smaller ME)
3. Variability in data (more variation → larger ME)

Effect on CI:
CI = (point estimate) ± ME
Larger ME → wider CI → less precise estimate


# Question 16: How is Bayes' Theorem used in statistics, and what is its significance?

 - Bayes' Theorem updates prior probabilities with new evidence:
P(A|B) = [P(B|A) * P(A)] / P(B)

Uses in statistics:
- Bayesian inference: Update beliefs about parameters as data is observed
- Machine learning: Naive Bayes classifiers
- Decision making under uncertainty

Significance:
- Incorporates prior knowledge
- Provides full probability distributions rather than point estimates
- Allows for intuitive interpretation of probabilities


# Question 17: What is the Chi-square distribution, and when is it used?

 - The Chi-square (χ²) distribution is a continuous probability distribution of the sum of squared standard normal deviates. It's used for:
1. Goodness-of-fit tests
2. Tests of independence in contingency tables
3. Tests of homogeneity
4. Variance estimation (for normal data)

Properties:
- Right-skewed
- Shape depends on degrees of freedom
- Mean = df, Variance = 2df


# Question 18: What is the Chi-square goodness of fit test, and how is it applied?

 - The Chi-square goodness-of-fit test compares observed frequencies to expected frequencies to test if a categorical variable follows a hypothesized distribution.

Steps:
1. State H₀ (data follows specified distribution) and H₁
2. Calculate expected frequencies
3. Compute χ² = Σ[(O-E)²/E]
4. Compare to χ² distribution with (k-1) df
5. Make decision based on P-value

Applications:
- Check if dice are fair
- Test if observed genotype frequencies match expected
- Verify distribution assumptions


# Question 19: What is the F-distribution, and when is it used in hypothesis testing?

 - The F-distribution is a continuous probability distribution of the ratio of two independent chi-square variables. It's used for:
1. ANOVA (comparing multiple group means)
2. Comparing variances (F-test)
3. Regression analysis (testing overall model significance)

Properties:
- Right-skewed
- Two parameters: numerator and denominator degrees of freedom
- Mean = d2/(d2-2) for d2 > 2


# Question 20: What is an ANOVA test, and what are its assumptions?

 - ANOVA (Analysis of Variance) tests for differences among group means by analyzing variance components.

Assumptions:
1. Independence of observations
2. Normally distributed residuals
3. Homogeneity of variances (equal variances across groups)
4. Interval or ratio level dependent variable

Types:
1. One-way ANOVA: Single factor with ≥2 levels
2. Two-way ANOVA: Two factors
3. MANOVA: Multiple dependent variables


# Question 21: What are the different types of ANOVA tests?

 - 1. One-way ANOVA:
   - Tests effect of one categorical independent variable on a continuous dependent variable
   - Compares means across 3+ groups

2. Two-way ANOVA:
   - Tests effects of two categorical independent variables and their interaction
   - Can assess main effects and interaction effects

3. Repeated Measures ANOVA:
   - For dependent samples/within-subjects designs
   - Same subjects measured under different conditions

4. MANOVA (Multivariate ANOVA):
   - Extends ANOVA to multiple dependent variables
   - Tests for overall differences across groups


# Question 22: What is the F-test, and how does it relate to hypothesis testing?

 - The F-test is a statistical test that uses the F-distribution to compare variances or test overall significance in models.

Common uses:
1. Comparing two variances (H₀: σ1² = σ2²)
2. Testing overall significance in regression (H₀: all coefficients = 0)
3. ANOVA (H₀: all group means are equal)

F-statistic calculation:
F = (explained variance) / (unexplained variance)
= (MS between groups) / (MS within groups) for ANOVA

Decision rule:
Reject H₀ if F > F-critical or P-value < α

```python
#2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

print("2. Discrete uniform distribution and PMF:")
low_val = 1
high_val = 6
num_samples = 1000
discrete_uniform_data = np.random.randint(low_val, high_val + 1, num_samples)

# Calculate PMF
unique, counts = np.unique(discrete_uniform_data, return_counts=True)
pmf = counts / num_samples

plt.figure(figsize=(8, 5))
plt.bar(unique, pmf, color='skyblue', edgecolor='black')
plt.title('Probability Mass Function of a Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(unique)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print("\n")
```

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

print("3. PDF of a Bernoulli distribution:")
def bernoulli_pdf(p, k):
    """
    Calculates the probability distribution function (PDF) of a Bernoulli distribution.
    p: probability of success (between 0 and 1)
    k: outcome (0 for failure, 1 for success)
    """
    if k == 1:
        return p
    elif k == 0:
        return 1 - p
    else:
        return 0

p_success = 0.7
print(f"PDF for success (k=1) with p={p_success}: {bernoulli_pdf(p_success, 1)}")
print(f"PDF for failure (k=0) with p={p_success}: {bernoulli_pdf(p_success, 0)}\n")
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
print("10. Implement Central Limit Theorem (CLT) for a non-normal distribution:")
# Using an exponential distribution as a non-normal example
from scipy.stats import expon

num_samples_clt = 1000 # Number of sample means
sample_size_clt = 30 # Size of each sample
lambda_exp = 0.5 # Rate parameter for exponential distribution

sample_means = []
for _ in range(num_samples_clt):
    sample = expon.rvs(scale=1/lambda_exp, size=sample_size_clt)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
plt.title(f'Histogram of Sample Means (CLT demonstration - Exponential Distribution)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print("As predicted by CLT, the distribution of sample means approaches a normal distribution, even if the original distribution is non-normal.\n")
```

```python
# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
print("14. Calculate Z-score for a data point and compare:")
data_point = 75
data_mean = 70
data_std = 5

z_score_single = (data_point - data_mean) / data_std
print(f"Data point: {data_point}, Mean: {data_mean}, Standard Deviation: {data_std}")
print(f"Calculated Z-score: {z_score_single:.2f}")

# Visualize the Z-score on a standard normal curve
x_norm = np.linspace(-4, 4, 1000)
pdf_norm = norm.pdf(x_norm, loc=0, scale=1)

plt.figure(figsize=(8, 5))
plt.plot(x_norm, pdf_norm, color='blue', label='Standard Normal PDF')
plt.axvline(z_score_single, color='red', linestyle='--', label=f'Z-score = {z_score_single:.2f}')
plt.fill_between(x_norm, 0, pdf_norm, where=(x_norm <= z_score_single), color='skyblue', alpha=0.4, label='P(Z <= Z-score)')
plt.title('Z-score on Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
print("The red dashed line indicates the calculated Z-score on the standard normal distribution.\n")

```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.
print("15. Hypothesis testing using Z-statistics:")
from statsmodels.stats.weightstats import ztest

# Sample data
sample_data_ht = np.array([68, 70, 72, 71, 69, 73, 70, 74, 67, 71])
population_mean_ht = 70
alpha_ht = 0.05 # Significance level

z_statistic, p_value_ht = ztest(sample_data_ht, value=population_mean_ht)

print(f"Sample Mean: {np.mean(sample_data_ht):.2f}")
print(f"Population Mean (Hypothesized): {population_mean_ht}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value_ht:.4f}")
print(f"Significance Level (alpha): {alpha_ht}")

if p_value_ht < alpha_ht:
    print("Reject the null hypothesis. There is significant evidence to suggest the sample mean is different from the population mean.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest the sample mean is different from the population mean.")
print("\n")

```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.
print("16. Confidence interval for a dataset:")
from scipy import stats

data_ci = np.array([25, 28, 30, 27, 26, 29, 31, 28, 27, 30])
confidence_level = 0.95

n_ci = len(data_ci)
mean_ci = np.mean(data_ci)
std_err_ci = stats.sem(data_ci) # Standard error of the mean

# For a 95% confidence interval, t-score with n-1 degrees of freedom
# For large n, Z-score can be used, but t-score is more general.
# Using t-distribution for generality
degrees_freedom = n_ci - 1
t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_freedom)
margin_of_error_ci = t_critical * std_err_ci
confidence_interval = (mean_ci - margin_of_error_ci, mean_ci + margin_of_error_ci)

print(f"Dataset: {data_ci}")
print(f"Sample Mean: {mean_ci:.2f}")
print(f"Standard Error of Mean: {std_err_ci:.2f}")
print(f"Confidence Level: {confidence_level * 100}%")
print(f"Margin of Error: {margin_of_error_ci:.2f}")
print(f"Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
print(f"Interpretation: We are {confidence_level * 100}% confident that the true population mean lies within the interval ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f}).\n")
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
print("17. CI for mean of normally distributed data:")
np.random.seed(42) # for reproducibility
data_normal_ci = np.random.normal(loc=100, scale=15, size=50) # Mean 100, Std Dev 15, 50 samples
confidence_level_normal = 0.99

n_normal_ci = len(data_normal_ci)
mean_normal_ci = np.mean(data_normal_ci)
std_err_normal_ci = stats.sem(data_normal_ci)

degrees_freedom_normal = n_normal_ci - 1
t_critical_normal = stats.t.ppf((1 + confidence_level_normal) / 2, degrees_freedom_normal)
margin_of_error_normal = t_critical_normal * std_err_normal_ci
confidence_interval_normal = (mean_normal_ci - margin_of_error_normal, mean_normal_ci + margin_of_error_normal)

print(f"Generated Normal Data Sample Mean: {mean_normal_ci:.2f}")
print(f"Standard Error of Mean: {std_err_normal_ci:.2f}")
print(f"Confidence Level: {confidence_level_normal * 100}%")
print(f"Margin of Error: {margin_of_error_normal:.2f}")
print(f"Confidence Interval for Mean: ({confidence_interval_normal[0]:.2f}, {confidence_interval_normal[1]:.2f})")
print(f"Interpretation: We are {confidence_level_normal * 100}% confident that the true population mean of the normal distribution from which this data was drawn lies within the interval ({confidence_interval_normal[0]:.2f}, {confidence_interval_normal[1]:.2f}).\n")

```

```python
# 24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
print("24. Calculate skewness and kurtosis of normal distribution data:")
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data) # Fisher's kurtosis (excess kurtosis)

    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (Fisher, excess): {data_kurtosis:.4f}")
    if -0.5 <= data_skewness <= 0.5:
        print("Interpretation (Skewness): The distribution is approximately symmetrical.")
    elif data_skewness > 0.5:
        print("Interpretation (Skewness): The distribution is positively (right) skewed.")
    else:
        print("Interpretation (Skewness): The distribution is negatively (left) skewed.")

    if -0.5 <= data_kurtosis <= 0.5:
        print("Interpretation (Kurtosis): The distribution is mesokurtic (similar to normal distribution).")
    elif data_kurtosis > 0.5:
        print("Interpretation (Kurtosis): The distribution is leptokurtic (heavy tails, more peaked).")
    else:
        print("Interpretation (Kurtosis): The distribution is platykurtic (light tails, less peaked).")

np.random.seed(42)
normal_data_skk = np.random.normal(loc=0, scale=1, size=1000)
print("For a Normal Distribution:")
calculate_skewness_kurtosis(normal_data_skk)
print("\n")

# Example for a skewed distribution (e.g., exponential)
print("For an Exponential Distribution (example of skewed data):")
exponential_data_skk = np.random.exponential(scale=1, size=1000)
calculate_skewness_kurtosis(exponential_data_skk)
print("\n")

# Example for a leptokurtic distribution (e.g., Laplace)
print("For a Laplace Distribution (example of leptokurtic data):")
from scipy.stats import laplace
laplace_data_skk = laplace.rvs(loc=0, scale=1, size=1000)
calculate_skewness_kurtosis(laplace_data_skk)
print("\n")
```

```python
# Practical Part - 2

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats.weightstats import ztest, ttest_ind

# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
print("1. Z-test for comparing sample mean to known population mean:")
# Known population parameters
population_mean_known = 50
population_std_known = 5
# Sample data
sample_data_z = np.array([48, 52, 51, 49, 53, 50, 47, 52, 50, 54])
alpha_z = 0.05 # Significance level

# Perform Z-test
# statsmodels.stats.weightstats.ztest requires population standard deviation if given,
# otherwise it calculates sample std dev. Here we assume population std is known.
# For simplicity, we'll manually calculate Z-score if population std is known and sample size is large enough.
# If using ztest from statsmodels, it usually implies large sample size for sample std dev to approximate pop std.
# Let's use the statsmodels ztest which is robust.
# For this specific question, if population std is *known*, the formula is (sample_mean - pop_mean) / (pop_std / sqrt(n))

sample_mean_z = np.mean(sample_data_z)
n_z = len(sample_data_z)
# Calculate Z-statistic manually as population std is known
z_statistic_manual = (sample_mean_z - population_mean_known) / (population_std_known / np.sqrt(n_z))
p_value_manual = 2 * (1 - stats.norm.cdf(abs(z_statistic_manual))) # Two-tailed p-value

print(f"Sample Data: {sample_data_z}")
print(f"Sample Mean: {sample_mean_z:.2f}")
print(f"Population Mean (Hypothesized): {population_mean_known}")
print(f"Known Population Standard Deviation: {population_std_known}")
print(f"Calculated Z-statistic: {z_statistic_manual:.4f}")
print(f"P-value: {p_value_manual:.4f}")
print(f"Significance Level (alpha): {alpha_z}")

if p_value_manual < alpha_z:
    print("Conclusion: Reject the null hypothesis. There is significant evidence that the sample mean is different from the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is not enough evidence to suggest the sample mean is different from the population mean.")
print("\n")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
print("4. Two-tailed Z-test with visualization of decision region:")
# Re-using the data from question 1 or 3 for demonstration. Let's use Q1's data.
sample_data_two_tail = np.array([48, 52, 51, 49, 53, 50, 47, 52, 50, 54])
population_mean_two_tail = 50
population_std_two_tail = 5 # Known population std
alpha_two_tail = 0.05

sample_mean_two_tail = np.mean(sample_data_two_tail)
n_two_tail = len(sample_data_two_tail)
z_statistic_two_tail = (sample_mean_two_tail - population_mean_two_tail) / (population_std_two_tail / np.sqrt(n_two_tail))
p_value_two_tail = 2 * (1 - stats.norm.cdf(abs(z_statistic_two_tail)))

print(f"Z-statistic: {z_statistic_two_tail:.4f}")
print(f"P-value: {p_value_two_tail:.4f}")

# Visualization
x_two_tail = np.linspace(-4, 4, 1000)
pdf_two_tail = stats.norm.pdf(x_two_tail, 0, 1) # Standard normal distribution

# Critical Z-values for a two-tailed test
z_critical_lower = stats.norm.ppf(alpha_two_tail / 2)
z_critical_upper = stats.norm.ppf(1 - alpha_two_tail / 2)

plt.figure(figsize=(10, 6))
plt.plot(x_two_tail, pdf_two_tail, color='blue', label='Standard Normal PDF')

# Shade rejection regions
x_reject_lower = x_two_tail[x_two_tail < z_critical_lower]
y_reject_lower = pdf_two_tail[x_two_tail < z_critical_lower]
plt.fill_between(x_reject_lower, 0, y_reject_lower, color='red', alpha=0.3, label='Rejection Region (alpha/2)')

x_reject_upper = x_two_tail[x_two_tail > z_critical_upper]
y_reject_upper = pdf_two_tail[x_two_tail > z_critical_upper]
plt.fill_between(x_reject_upper, 0, y_reject_upper, color='red', alpha=0.3)

# Plot calculated Z-statistic
plt.axvline(z_statistic_two_tail, color='green', linestyle='--', label=f'Calculated Z-statistic = {z_statistic_two_tail:.2f}')
plt.axvline(z_critical_lower, color='grey', linestyle=':', label=f'Critical Z = {z_critical_lower:.2f}')
plt.axvline(z_critical_upper, color='grey', linestyle=':', label=f'Critical Z = {z_critical_upper:.2f}')

plt.title('Two-tailed Z-test: Decision Region')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

if abs(z_statistic_two_tail) > z_critical_upper: # Or p_value < alpha
    print("Conclusion: The calculated Z-statistic falls in the rejection region. Reject the null hypothesis.")
else:
    print("Conclusion: The calculated Z-statistic falls outside the rejection region. Fail to reject the null hypothesis.")
print("\n")
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
print("5. Calculate and visualize Type 1 and Type 2 errors:")
def visualize_errors(mu0, mu1, sigma, n, alpha):
    """
    Visualizes Type I and Type II errors for a one-sample Z-test.
    mu0: Null hypothesis mean
    mu1: True population mean (for calculating Beta/Type II error)
    sigma: Population standard deviation
    n: Sample size
    alpha: Significance level (Type I error probability)
    """
    std_error = sigma / np.sqrt(n)

    # Type I error (alpha)
    # Critical value under H0 (Z-score)
    z_critical = stats.norm.ppf(1 - alpha)
    # Critical sample mean value
    x_critical = mu0 + z_critical * std_error

    # Type II error (beta)
    # Probability of failing to reject H0 when H1 is true
    beta = stats.norm.cdf((x_critical - mu1) / std_error)

    print(f"Null Hypothesis Mean (mu0): {mu0}")
    print(f"True Population Mean (mu1): {mu1}")
    print(f"Population Standard Deviation (sigma): {sigma}")
    print(f"Sample Size (n): {n}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Calculated Type I Error (alpha): {alpha:.4f}")
    print(f"Calculated Type II Error (beta): {beta:.4f}")
    print(f"Power (1 - beta): {1 - beta:.4f}")

    # Visualization
    x_range = np.linspace(min(mu0, mu1) - 4 * std_error, max(mu0, mu1) + 4 * std_error, 500)
    pdf_h0 = stats.norm.pdf(x_range, mu0, std_error)
    pdf_h1 = stats.norm.pdf(x_range, mu1, std_error)

    plt.figure(figsize=(10, 6))
    plt.plot(x_range, pdf_h0, label='Distribution under H0', color='blue')
    plt.plot(x_range, pdf_h1, label='Distribution under H1 (True)', color='green')

    # Shade Type I error (alpha)
    x_alpha = x_range[x_range > x_critical]
    y_alpha = pdf_h0[x_range > x_critical]
    plt.fill_between(x_alpha, 0, y_alpha, color='red', alpha=0.3, label='Type I Error (alpha)')

    # Shade Type II error (beta)
    x_beta = x_range[x_range < x_critical]
    y_beta = pdf_h1[x_range < x_critical]
    plt.fill_between(x_beta, 0, y_beta, color='orange', alpha=0.3, label='Type II Error (beta)')

    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical Value (X_crit) = {x_critical:.2f}')
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Example usage
visualize_errors(mu0=100, mu1=102, sigma=10, n=30, alpha=0.05)
visualize_errors(mu0=100, mu1=101, sigma=10, n=100, alpha=0.01) # Smaller alpha, larger n
print("\n")
```

```python
# 8. Simulate data and perform both Z-test and T-test, then compare the results using Python.
print("8. Simulate data and compare Z-test vs. T-test:")
np.random.seed(42)
# Data for Z-test (large sample, assumed known population std for this comparison)
sample_data_z_comp = np.random.normal(loc=80, scale=10, size=100) # Large sample
population_mean_z_comp = 82
population_std_z_comp = 10 # Known population std

# Z-test
z_statistic_comp = (np.mean(sample_data_z_comp) - population_mean_z_comp) / (population_std_z_comp / np.sqrt(len(sample_data_z_comp)))
p_value_z_comp = 2 * (1 - stats.norm.cdf(abs(z_statistic_comp)))

print("--- Z-test Results ---")
print(f"Sample Mean: {np.mean(sample_data_z_comp):.2f}")
print(f"Hypothesized Population Mean: {population_mean_z_comp}")
print(f"Z-statistic: {z_statistic_comp:.4f}")
print(f"P-value: {p_value_z_comp:.4f}")
print(f"Conclusion (alpha=0.05): {'Reject H0' if p_value_z_comp < 0.05 else 'Fail to Reject H0'}")

print("\n--- T-test Results (using the same data) ---")
# T-test (usually preferred when population std is unknown or sample size is small)
t_statistic_comp, p_value_t_comp = stats.ttest_1samp(sample_data_z_comp, population_mean_z_comp)

print(f"T-statistic: {t_statistic_comp:.4f}")
print(f"P-value: {p_value_t_comp:.4f}")
print(f"Conclusion (alpha=0.05): {'Reject H0' if p_value_t_comp < 0.05 else 'Fail to Reject H0'}")

print("\nComparison:")
print("When the sample size is large, the Z-test and T-test results (especially p-values) tend to be very similar because the T-distribution approaches the Normal distribution.")
print("The Z-test strictly requires knowing the population standard deviation, whereas the T-test uses the sample standard deviation, making it more widely applicable in practice.\n")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
print("9. Confidence interval for a sample mean and its significance:")
def calculate_confidence_interval(data, confidence_level=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data) # Standard error of the mean

    # Use t-distribution for confidence interval of mean unless population std is known and n is very large
    degrees_freedom = n - 1
    t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_freedom)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Data (first 5 values): {data[:5]}...")
    print(f"Sample Size (n): {n}")
    print(f"Sample Mean: {mean:.2f}")
    print(f"Standard Error of Mean: {std_err:.4f}")
    print(f"Confidence Level: {confidence_level * 100}%")
    print(f"Degrees of Freedom: {degrees_freedom}")
    print(f"T-critical value: {t_critical:.4f}")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print("Significance: A confidence interval provides a range of plausible values for an unknown population parameter (e.g., the true population mean).")
    print(f"A {confidence_level * 100}% confidence interval means that if we were to take many samples and construct a confidence interval for each, approximately {confidence_level * 100}% of these intervals would contain the true population mean.")
    print("It quantifies the uncertainty associated with estimating a population parameter based on a sample.\n")

# Example usage
np.random.seed(10)
sample_data_ci_func = np.random.normal(loc=65, scale=8, size=40)
calculate_confidence_interval(sample_data_ci_func, confidence_level=0.95)
calculate_confidence_interval(sample_data_ci_func, confidence_level=0.99)
print("\n")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
print("10. Calculate margin of error:")
sample_data_moe = np.array([12, 15, 11, 14, 16, 13, 15, 12, 14, 13, 17, 10, 16, 15, 14])
confidence_level_moe = 0.90 # 90% confidence

n_moe = len(sample_data_moe)
mean_moe = np.mean(sample_data_moe)
std_dev_moe = np.std(sample_data_moe, ddof=1) # Sample standard deviation (n-1 for unbiased)
std_err_moe = std_dev_moe / np.sqrt(n_moe)

degrees_freedom_moe = n_moe - 1
t_critical_moe = stats.t.ppf((1 + confidence_level_moe) / 2, degrees_freedom_moe)

margin_of_error_calc = t_critical_moe * std_err_moe

print(f"Sample Data: {sample_data_moe}")
print(f"Sample Size (n): {n_moe}")
print(f"Sample Mean: {mean_moe:.2f}")
print(f"Sample Standard Deviation: {std_dev_moe:.2f}")
print(f"Standard Error of Mean: {std_err_moe:.2f}")
print(f"Confidence Level: {confidence_level_moe * 100}%")
print(f"T-critical value: {t_critical_moe:.4f}")
print(f"Calculated Margin of Error: {margin_of_error_calc:.2f}")
print("The margin of error indicates the maximum likely difference between the sample estimate and the true population parameter for a given confidence level.\n")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
print("11. Bayesian inference using Bayes' Theorem:")
# Example: Drug test for a disease
# P(D) = Probability of having the disease (prior probability)
# P(Pos|D) = Probability of a positive test given the person has the disease (sensitivity)
# P(Neg|not D) = Probability of a negative test given the person does not have the disease (specificity)
# We want to find P(D|Pos) = Probability of having the disease given a positive test (posterior probability)

# Prior probability: Prevalence of the disease in the population
P_D = 0.01  # 1% of the population has the disease

# Likelihoods (from test characteristics)
P_Pos_given_D = 0.95  # Sensitivity: 95% chance of positive test if you have the disease
P_Neg_given_not_D = 0.90 # Specificity: 90% chance of negative test if you don't have the disease

# Calculate other necessary probabilities
P_not_D = 1 - P_D
P_Pos_given_not_D = 1 - P_Neg_given_not_D # False positive rate: 10%

# Calculate the marginal likelihood P(Pos) using the law of total probability
# P(Pos) = P(Pos|D) * P(D) + P(Pos|not D) * P(not D)
P_Pos = (P_Pos_given_D * P_D) + (P_Pos_given_not_D * P_not_D)

# Apply Bayes' Theorem
# P(D|Pos) = (P(Pos|D) * P(D)) / P(Pos)
P_D_given_Pos = (P_Pos_given_D * P_D) / P_Pos

print("--- Bayesian Inference Example: Drug Test ---")
print(f"Prior Probability of Disease P(D): {P_D:.4f}")
print(f"Probability of Positive Test given Disease P(Pos|D): {P_Pos_given_D:.4f}")
print(f"Probability of Positive Test given No Disease P(Pos|not D): {P_Pos_given_not_D:.4f}")
print(f"Marginal Probability of Positive Test P(Pos): {P_Pos:.4f}")
print(f"Posterior Probability of Disease given Positive Test P(D|Pos): {P_D_given_Pos:.4f}")

print("\nExplanation:")
print("Bayes' Theorem provides a way to update the probability of a hypothesis (having the disease) as more evidence (a positive test result) becomes available.")
print("1. **Prior Probability (P_D)**: Our initial belief about the probability of the disease before seeing any evidence.")
print("2. **Likelihood (P_Pos_given_D and P_Pos_given_not_D)**: The probability of observing the evidence (positive test) under different hypotheses (having or not having the disease).")
print("3. **Marginal Likelihood (P_Pos)**: The overall probability of observing the evidence, accounting for all possible hypotheses.")
print("4. **Posterior Probability (P_D_given_Pos)**: The updated probability of the hypothesis (having the disease) after considering the new evidence. This is the core of Bayesian inference.")
print("In this example, despite a high test sensitivity, the relatively low prior probability of the disease leads to a posterior probability that is still quite low, highlighting the importance of prior beliefs in Bayesian analysis.\n")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python.
print("12. Chi-square test for independence:")
from scipy.stats import chi2_contingency

# Example: Relationship between gender and preference for a certain product
# Rows: Gender (Male, Female)
# Columns: Product Preference (Prefers A, Prefers B, Prefers C)
observed_data = np.array([
    [50, 30, 20],  # Male counts
    [40, 60, 50]   # Female counts
])

chi2_statistic, p_value_chi2, degrees_freedom_chi2, expected_frequencies_chi2 = chi2_contingency(observed_data)

print("Observed Frequencies:\n", observed_data)
print(f"Chi-square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value_chi2:.4f}")
print(f"Degrees of Freedom: {degrees_freedom_chi2}")
print("Expected Frequencies (under independence):\n", expected_frequencies_chi2)
alpha_chi2 = 0.05

if p_value_chi2 < alpha_chi2:
    print("Conclusion: Reject the null hypothesis. There is a significant association (dependence) between gender and product preference.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant association (independence) between gender and product preference.")
print("\n")
```

```python
# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
print("14. Chi-square goodness-of-fit test:")
# Example: Testing if dice rolls are fair
# Observed frequencies of rolling 1, 2, 3, 4, 5, 6
observed_rolls = np.array([20, 18, 22, 25, 17, 18])
total_rolls = np.sum(observed_rolls)

# Expected frequencies for a fair dice (uniform distribution)
num_categories = len(observed_rolls)
expected_rolls = np.full(num_categories, total_rolls / num_categories)

print(f"Observed Frequencies of Dice Rolls: {observed_rolls}")
print(f"Expected Frequencies (Fair Dice): {expected_rolls}")

# Perform chi-square goodness-of-fit test
# H0: The observed distribution matches the expected distribution (e.g., dice is fair)
# H1: The observed distribution does not match the expected distribution (e.g., dice is biased)
chi2_gof_statistic, p_value_gof = stats.chisquare(f_obs=observed_rolls, f_exp=expected_rolls)

print(f"Chi-square Goodness-of-Fit Statistic: {chi2_gof_statistic:.4f}")
print(f"P-value: {p_value_gof:.4f}")
alpha_gof = 0.05

if p_value_gof < alpha_gof:
    print("Conclusion: Reject the null hypothesis. The observed dice rolls significantly differ from what would be expected from a fair dice.")
else:
    print("Conclusion: Fail to reject the null hypothesis. The observed dice rolls are consistent with a fair dice.")
print("\n")
```
